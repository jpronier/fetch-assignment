```python
import pandas as pd
import numpy as np
```

### Users

```python
# Reading users data
users = pd.read_json("../data/users.json", lines=True)
users.fillna(np.nan, inplace=True)
users.head()
```

```python
# Processing users data
# Extract the id from the '_id' column
users["user_id"] = users["_id"].apply(
    lambda x: x["$oid"] if isinstance(x, dict) and "$oid" in x else np.nan
)
users["createdDate"] = users["createdDate"].apply(
    lambda x: x["$date"] if isinstance(x, dict) and "$date" in x else np.nan
)
users["lastLogin"] = users["lastLogin"].apply(
    lambda x: x["$date"] if isinstance(x, dict) and "$date" in x else np.nan
)

# dropping column _id
users = users[
    ["user_id", "role", "createdDate", "lastLogin", "active", "state", "signUpSource"]
]

# handling datetime values
users["createdDate"] = pd.to_datetime(users["createdDate"], unit="ms")
users["lastLogin"] = pd.to_datetime(users["lastLogin"], unit="ms")

users.head()
```

### Receipts

```python
# Reading receipts data
receipts = pd.read_json("../data/receipts.json", lines=True)
receipts.head()
```

```python
# Processing receipts data
# Extract the id from the '_id' column
receipts["receipt_id"] = receipts["_id"].apply(
    lambda x: x["$oid"] if isinstance(x, dict) and "$oid" in x else np.nan
)

# Convert date fields
date_fields = [
    "dateScanned",
    "purchaseDate",
    "modifyDate",
    "createDate",
    "pointsAwardedDate",
    "finishedDate",
]
receipts[date_fields] = receipts[date_fields].map(
    lambda x: x["$date"] if isinstance(x, dict) and "$date" in x else np.nan
)
receipts[date_fields] = receipts[date_fields].apply(pd.to_datetime, unit="ms")

receipts = receipts[
    [
        "receipt_id",
        "userId",
        "bonusPointsEarned",
        "bonusPointsEarnedReason",
        "createDate",
        "dateScanned",
        "finishedDate",
        "modifyDate",
        "pointsAwardedDate",
        "purchaseDate",
        "purchasedItemCount",
        "rewardsReceiptItemList",
        "rewardsReceiptStatus",
        "totalSpent",
    ]
]

# Handle any remaining NaN values
# receipts.fillna(np.nan, inplace=True)

receipts.head()
```

```python
w_receipts = receipts
```

```python
item_df = w_receipts[["receipt_id", "rewardsReceiptItemList"]].explode(
    "rewardsReceiptItemList"
)

# resetting index
item_df.reset_index(drop=True, inplace=True)

# normalizing
item_df = pd.concat(
    [item_df["receipt_id"], pd.json_normalize(item_df["rewardsReceiptItemList"])],
    axis=1,
)

item_df.head()
```

```python
# Process brands table similar to users and receipts

# Convert _id to string
brands["brand_id"] = brands["_id"].apply(lambda x: x["$oid"])

# Convert cpg.$id to string
brands["cpg_id"] = brands["cpg"].apply(lambda x: x["$id"]["$oid"])

# Convert topBrand to boolean
brands["topBrand"] = brands["topBrand"].astype(bool)

brands = brands[
    [
        "brand_id",
        "cpg_id",
        "barcode",
        "category",
        "categoryCode",
        "name",
        "topBrand",
        "brandCode",
    ]
]

brands.head()
```
